```python
#Normalizing the data

df = pd.DataFrame(all_stock_data)

# Select columns to normalize (excluding non-numeric columns)
columns_to_normalize = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']

# Min-Max scaling
for column in columns_to_normalize:
    df[column + '_Normalized'] = (df[column] - df[column].min()) / (df[column].max() - df[column].min())

print(df)
```

```python
import pandas as pd

def calculate_rsi(data, window=14):
    delta = data['Close'].diff(1)
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()

    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Assuming your DataFrame is named 'df' and contains a column named 'Close' for closing prices
# Also assuming the data is sorted from oldest to newest

# Calculate RSI
df['RSI'] = calculate_rsi(df)

# Print the DataFrame to see the new 'RSI' column
print(df)
```
